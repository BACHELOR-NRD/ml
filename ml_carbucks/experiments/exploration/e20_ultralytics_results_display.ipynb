```python
import pandas as pd
import matplotlib.pyplot as plt

def display_ultralytics_progress(csv_path):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(csv_path)
    
    # Display the DataFrame
    print(df.columns)

    # plot "epoch" vs "metrics/mAP50(B)" and secondary y-axis "metrics/mAP50-95(B)"
    y_secondaries = [
        "val/cls_loss",
        "val/dfl_loss",
    ]
    df.plot(
        x="epoch", 
        y=["metrics/mAP50(B)", "metrics/mAP50-95(B)", *y_secondaries],
        secondary_y=[*y_secondaries],
    )

display_ultralytics_progress("/home/bachelor/ml-carbucks/optuna/hyper/custom_params/checkpoints/debug/debug_YoloUltralyticsAdapter_trial_3/results.csv")
```
